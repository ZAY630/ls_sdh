```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import pytz
```

```python
df = pd.read_csv("SW.csv")
df.head()
```

```python
print(df.columns)
```

```python
# Reorganize column name
prefix_to_remove = '/FS5/7200/SDH.SW.'
suffix_to_remove = '/presentValue'

df.columns = [col.replace(prefix_to_remove, '') if col.startswith(prefix_to_remove) else col for col in df.columns]
df.columns = [col.replace(suffix_to_remove, '') if col.endswith(suffix_to_remove) else col for col in df.columns]
df.columns = [col.split('/')[0] for col in df.columns]

df.head()
```

```python
print(df.columns)
```

```python
# change unit to kW
df.iloc[:, 1:] = df.iloc[:, 1:]/1000

# change datetime to datetime object
df['datetime'] = pd.to_datetime(df['datetime'])
df['datetime'] = df['datetime'].dt.tz_localize('America/Los_Angeles', ambiguous=True)
```

```python
# split into MSA and MSB dataframe
# Define the keywords
keywords = ['MSA.', 'MSB.']

# Identify columns containing the keywords
columns_with_keywords = {keyword: [col for col in df.columns if keyword in col] for keyword in keywords}

print("Columns with Keywords:")
print(columns_with_keywords)

# Create sub DataFrames based on identified columns
sub_dataframes = {}
for keyword, cols in columns_with_keywords.items():
    # Include the first column ('ID') and the columns with the keyword
    selected_cols = ['datetime'] + cols
    sub_dataframes[keyword] = df[selected_cols]

df_MSA = sub_dataframes['MSA.']
df_MSB = sub_dataframes['MSB.']
```

```python
# split into peak demand and power dataframe
# Define the keywords
keywords = ['DEMAND', 'PWR']

# Identify columns containing the keywords
columns_with_keywords = {keyword: [col for col in df_MSA.columns if keyword in col] for keyword in keywords}

print("Columns with Keywords:")
print(columns_with_keywords)

# Create sub DataFrames based on identified columns
sub_dataframes = {}
for keyword, cols in columns_with_keywords.items():
    # Include the first column ('ID') and the columns with the keyword
    selected_cols = ['datetime'] + cols
    sub_dataframes[keyword] = df[selected_cols]

df_MSA_DEMAND = sub_dataframes['DEMAND']
df_MSA_PWR = sub_dataframes['PWR']

# split into peak demand and power dataframe
# Define the keywords
keywords = ['DEMAND', 'PWR']

# Identify columns containing the keywords
columns_with_keywords = {keyword: [col for col in df_MSB.columns if keyword in col] for keyword in keywords}

print("Columns with Keywords:")
print(columns_with_keywords)

# Create sub DataFrames based on identified columns
sub_dataframes = {}
for keyword, cols in columns_with_keywords.items():
    # Include the first column ('ID') and the columns with the keyword
    selected_cols = ['datetime'] + cols
    sub_dataframes[keyword] = df[selected_cols]

df_MSB_DEMAND = sub_dataframes['DEMAND']
df_MSB_PWR = sub_dataframes['PWR']
```

```python
plt.figure(figsize=[20, 10])
for i in df_MSA_PWR.columns:
    if "datetime" not in i and "MAIN" not in i:
        plt.plot(df_MSA_PWR['datetime'], df_MSA_PWR[i], label = i)

plt.xlabel('Datetime', fontsize = 15)
plt.ylabel('Power consumption (kW)', fontsize = 15)
plt.title('Power consumption for substations MSA', fontsize = 18)
plt.legend()
plt.show()
```

```python
# Add a manual summation
# Specify the columns to be excluded
exclude_columns = ['datetime', 
                   'SDH.SW.MSA.MAIN.PWR REAL 3 P', 
                   'SDH.SW.MSA.MSB.TIE.PWR REAL 3P', 
                   'SDH.SW.MSA.MDC.PWR REAL 3 P', 
                   'SDH.SW.MSA.ATS-E2.PWR REAL 3 P', 
                   'SDH.SW.MSA.MSB.PWR REAL 3 P']

# Select the columns to be summed
columns_to_sum = [col for col in df_MSA_PWR.columns if col not in exclude_columns]

# Sum the selected columns row-wise and assign the result to a new column 'Sum'
df_MSA_PWR['SUM'] = df_MSA_PWR[columns_to_sum].sum(axis = 1)
df_MSA_PWR['diff'] = df_MSA_PWR['SUM'] - df_MSA_PWR['SDH.SW.MSA.MAIN.PWR REAL 3 P']

plt.figure(figsize=[20, 12])
plt.subplot(2, 1, 1)
plt.plot(df_MSA_PWR['datetime'], df_MSA_PWR['SUM'], label = "Mannual summation")
plt.plot(df_MSA_PWR['datetime'], df_MSA_PWR['SDH.SW.MSA.MAIN.PWR REAL 3 P'], label = "Main power meter")
plt.xlabel("Datetime", fontsize = 15)
plt.ylabel("Power consumption (kW)", fontsize = 15)
plt.legend()
plt.title("Comparison of mannual summation and main poweer meter measurements (MSA)", fontsize = 18)

plt.subplot(2, 1, 2)
plt.scatter(df_MSA_PWR['datetime'], df_MSA_PWR['diff'], label = "Difference", alpha = 0.5)
plt.xlabel("Datetime", fontsize = 15)
plt.ylabel("Power consumption (kW)", fontsize = 15)
plt.legend()
plt.title("Power difference (MSA)", fontsize = 18)
```

```python
print("MSA main panel include:", columns_to_sum)
print("mean error: ", np.mean(df_MSA_PWR['diff']), 'kW')
```

```python
plt.figure(figsize=[20, 10])
for i in df_MSB_PWR.columns:
    if "datetime" not in i and "MAIN" not in i:
        plt.plot(df_MSB_PWR['datetime'], df_MSB_PWR[i], label = i)

plt.xlabel('Datetime', fontsize = 15)
plt.ylabel('Power consumption (kW)', fontsize = 15)
plt.title('Power consumption for substations MSB', fontsize = 18)
plt.legend()
plt.show()
```

```python
# Add a manual summation
# Specify the columns to be excluded
exclude_columns = ['datetime', 'SDH.SW.MSB.MAIN.PWR REAL 3 P', 'SDH.SW.MSA.MSB.TIE.PWR REAL 3P', 'SDH.SW.MSA.MSB.PWR REAL 3 P', ]

# Select the columns to be summed
columns_to_sum = [col for col in df_MSB_PWR.columns if col not in exclude_columns]

# Sum the selected columns row-wise and assign the result to a new column 'Sum'
# MSA.ATS-E2 should be part of MSB
df_MSB_PWR['SUM'] = df_MSB_PWR[columns_to_sum].sum(axis = 1) + df_MSA_PWR['SDH.SW.MSA.ATS-E2.PWR REAL 3 P']
df_MSB_PWR['diff'] = df_MSB_PWR['SUM'] - df_MSB_PWR['SDH.SW.MSB.MAIN.PWR REAL 3 P']

plt.figure(figsize=[20, 12])
plt.subplot(2, 1, 1)
plt.plot(df_MSB_PWR['datetime'], df_MSB_PWR['SUM'], label = "Mannual summation")
plt.plot(df_MSB_PWR['datetime'], df_MSB_PWR['SDH.SW.MSB.MAIN.PWR REAL 3 P'], label = "Main power meter")
plt.xlabel("Datetime", fontsize = 15)
plt.ylabel("Power consumption (kW)", fontsize = 15)
plt.legend()
plt.title("Comparison of mannual summation and main poweer meter measurements (MSB)", fontsize = 18)

plt.subplot(2, 1, 2)
plt.scatter(df_MSB_PWR['datetime'], df_MSB_PWR['diff'], label = "Difference", alpha = 0.5)
plt.xlabel("Datetime", fontsize = 15)
plt.ylabel("Power consumption (kW)", fontsize = 15)
plt.legend()
plt.title("Power difference (MSB)", fontsize = 18)
```

### Load MOER data from WattTime API

```python
moer = pd.read_csv("moer.csv")
moer['point_time'] = pd.to_datetime(moer['point_time'])
moer['datetime'] = moer['point_time'].dt.tz_convert('America/Los_Angeles')
moer = moer.drop(columns='point_time')
moer.head()
```

```python
moer['moer'] = moer['value'] * 0.454
moer.set_index('datetime', inplace=True)
moer = moer.resample('15T').mean()
moer = moer.reset_index()
plt.figure(figsize = [20, 10])
plt.plot(moer['datetime'], moer['moer'])
plt.xlabel('Datetime', fontsize = 15)
plt.ylabel('MOER (g/kWh)', fontsize = 15)
plt.title('Marginal carbon emissions from WattTime', fontsize = 18)
```

```python
# Spot check for moer signals
plt.figure(figsize = [20, 20])
plt.subplot(3, 1, 1)
df = moer[moer['datetime'].dt.date == pd.to_datetime('2022-05-10')]
plt.plot(df['datetime'], df['moer'])
plt.xlabel('Datetime', fontsize = 15)
plt.ylabel('MOER (g/kWh)', fontsize = 15)
plt.title('Marginal carbon emissions from WattTime', fontsize = 18)

plt.subplot(3, 1, 2)
df = moer[moer['datetime'].dt.date == pd.to_datetime('2022-06-01')]
plt.plot(df['datetime'], df['moer'])
plt.xlabel('Datetime', fontsize = 15)
plt.ylabel('MOER (g/kWh)', fontsize = 15)
plt.title('Marginal carbon emissions from WattTime', fontsize = 18)

plt.subplot(3, 1, 3)
df = moer[moer['datetime'].dt.date == pd.to_datetime('2022-07-10')]
plt.plot(df['datetime'], df['moer'])
plt.xlabel('Datetime', fontsize = 15)
plt.ylabel('MOER (g/kWh)', fontsize = 15)
plt.title('Marginal carbon emissions from WattTime', fontsize = 18)
```
